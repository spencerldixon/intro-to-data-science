```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Magic helper for inline graphs in jupyter notebook

%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

## Functional vs Object Oriented: Functional

```python
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('My Graph')
plt.show()
```

```python
plt.subplot(1,2,1) # Number of rows, number of columns, plot we're referring to
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')

plt.show()
```

## Object Oriented

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Left, bottom, width, height
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title("My Title")
```

```python
fig = plt.figure(figsize=(5,5))

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Left, bottom, width, height
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y,'r')
axes2.plot(y,x)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[0].set_title("first plot")

axes[1].plot(y, x)
axes[1].set_title("second plot")

plt.tight_layout()
```

```python
axes
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, label="Things") # Must use labels for legends
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title("My Chart")

ax.legend(loc="best") # can use strings or location as a tuple
```

```python
fig.savefig('my_chart.png', dpi=300)
```

## Styling

```python
# setting colours
# line width and types
# Markers

z = np.random.rand(11)

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color="red", linewidth=1) # Can use strings, or hex
ax.plot(y, x, color="#0000FF", linewidth=3, alpha=0.5, linestyle="--")
ax.plot(y, z, color="green", linestyle=":", marker="o", markersize="5")
ax.plot
```

# Seaborn

statistical plotting on top of matplotlib

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30) # kernel density estimation
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.pairplot(tips, hue='sex')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
# x is categorical, y is numerical
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```
